## 1. Modeling Transcriptional Regulation: The Hill Function

In systems biology, the binding of a transcription factor (TF) to a promoter is commonly modeled using the **Hill function**. This function describes how the  activity of a promoter  depends on the concentration of its active transcription factor, $X^*$. Since the promoter allow the transcription of an mRNA which will be transcribed in a protein defined as Y,we can summarive this procces as:

rate of production of Y = $f(X^*)$

### Mathematical Formulation

More specifically the activity of the promoter can be described by this monotonic hill function: 

$$f(X^*)=\frac{\beta X^{*n}}{K^{n} + X^{*n}}$$

Where:
* $X^*$ it's the concentration of the active transcription factor
* K is the **activation coefficent**, it defines the concetration of active X needed to significantly activate the expression
* $\beta$ is the maximum expression of the promoter
* n govern the stepness of the function, the larger is n and the more step-like the function will be




```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
plt.rcParams.update({'font.size': 11, 'axes.labelsize': 12})

# 1. Define Hill normalized function
def hill_activation(x_ratio, n):
    """
    Calculate the normalized activity of the promoter
    """
    return (x_ratio**n) / (1 + x_ratio**n)


x_vals = np.linspace(0, 4, 500)


n_values = [1, 2, 4, 8]


plt.figure(figsize=(8, 5.5))

for n in n_values:
    y_vals = hill_activation(x_vals, n)
    #  n=1 (Michaelis-Menten), gradual response
    if n == 1:
        label_text = f'n = {n}'
        linestyle = '--'
    else:
        label_text = f'n = {n}'
        linestyle = '-'
        
    plt.plot(x_vals, y_vals, label=label_text, linestyle=linestyle, linewidth=2)

#evidentiate when X=k therefore X/K=1 and the activity of promoter is B/2 where B is Its maximum activity
plt.axvline(x=1.0, color='gray', linestyle=':', alpha=0.7)
plt.axhline(y=0.5, color='gray', linestyle=':', alpha=0.7)
plt.plot(1.0, 0.5, 'ro', label='K (Half-maximal activation)')

# labels of the graph
plt.title("Hill Activation Function (Normalized)", fontsize=14, fontweight='bold', pad=15)
plt.xlabel(r"Transcription Factor Concentration Ratio $\left(\frac{X^*}{K}\right)$")
plt.ylabel(r"Normalized Promoter Activity $\left(\frac{f(X^*)}{\beta}\right)$")
plt.xlim(0, 4)
plt.ylim(0, 1.05)
plt.legend(loc="lower right", frameon=True)
plt.tight_layout()

# show graph
plt.show()
```

From the graph below we can notice that when $X^*=K$ all the functions converges to one point which is the the **treshold** for activation of a target gene and correspond to the concentration of $X^*$ needed for $50\%$ maximal activation.
