# Aploid Models of Natural selection
Consider a population of two type of Individuals (A and a), A produces only A offspring and a produces only a offspring.
This scenarion describes the case of a haploid population with two alleles (A and a) at one genetic locus.
They reproduces with a reproductive factor denoted by W and we allow that term to be different between the 2 species therefore exist $W_a$ and $W_b$.

Assuming an *exponential population growth* we can describe the model using discrete time step using the following equations:
$$ n'(t) = n(t) + bn(t)= (1+b)n(t)\space\space (after\space births) $$
$$ n''(t) = n'(t) - dn'(t) = (1-d)n'(t)\space\space (after\space deaths) $$
Where $b$ is the percapita number of births and $d$ is the fraction of the population that dies, from this equations we can infer how the population change after a discrete time step, this is event is described by the following equation:
$$ n(t+1) = n''(t) = (1-d)(1+b)n(t)$$
from that equation we define $W=(1-d)(1+b)$ as the *reproductive factor* which is the number of surving individual per parent.

Once understand where reproductive factor come from we back to the initial Aploid model assuming it follow an exponeniantial growth population we define the following equations for the two type of populations:
$$n_A(t+1) = W_An_A(t) $$
$$n_a(t+1) = W_an_a(t) $$

Most evolutionary models focus on the frequency of each type rather than their absolute numbers. Tipically, p and q are used to denote the proportions of A and a alleles, given that we define numerically theese proportions:
$$p=\frac{n_A}{n_A+n_a}$$
$$q=\frac{n_a}{n_A+n_a}$$

Since p+q=1 we can use only one of them to define the other (e.g q=1-p).

By plugging theese equations to the previous ones we can obtain the allele frequency at time t+1:
$$p(t+1) = \frac{n_A(t+1)}{n_A(t+1)+n_a(t+1)}$$
$$\space\space\space =\frac{W_An_A(t)}{W_An_A(t)+W_an_a(t)}$$

To obtain a recursion equation that can predict the dynamics of the allele frequency over the course of several generations we need to write the previous equation in terms of p(t). To achieve that we divide the top and bottom by $(n_A+n_a)$, then each term involves only p or q :
$$p(t+1)=\frac{W_Ap(t)}{W_Ap(t)+W_a(1-p(t))}$$

In this model the numbers of surviving individuals per parent for the two alleles are $W_A$ and $W_a$. Since the equation above contains both of these parameters we can reduce the number or parameters dividing the top and the bottom by $W_a$ which produces an equation with only one parameter:
$$ p(t+1) = \frac{V_A p(t)}{V_A p(t) + (1-p(t))}$$

Where $V_A =\frac{W_A}{W_a} $

The fact that the dynamics of allele frequency depend only on the relative fitnesses of the two alleles reveals important information about the causes of evolutionary change. This a key insight: **evolutionary change can be studied without reference to the ecological context of a population as long as the ecological context affects the reproductive output of each allele by the same factor.**
