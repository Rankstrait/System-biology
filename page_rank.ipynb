```python
import numpy as np
```

```python
M = np.array(([0, 0, 1], #get stucked to state 3
             [0.5, 0, 0.5],
             [0.3, 0.3, 0.4]))
```

```python
def pagerank(M, d, tol=1e-3):
    n = M.shape[0]
    v =np.ones(n)/n
    
    M_hat = (1-d)*M + d/n
    for i in range(100):
        v = M_hat @ v
    
    return v
```

```python
pagerank(M.T, .1)
```
